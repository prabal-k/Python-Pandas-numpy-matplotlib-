```python
df3=df2.copy()
```

```python
#Mean imputation ie filling the null values with mean
df3['Age'] = df2['Age'].fillna(df2['Age'].mean())
```

```python
df3.info()
```

```python
# to find the outlier of age column
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
plt.boxplot(df3['Age'],vert=False)
plt.ylabel('Variable')
plt.xlabel('Age')
plt.title('Box Plot')
plt.show()
```

```python
#Used normal distribution (look at the above histogram of age as there is high difference between ages)
#calculate summary statistics
mean = df3['Age'].mean()
std  = df3['Age'].std()
print('Mean :',mean)
print('std :',std)

# Calculate the lower and upper bounds
lower_bound = mean - std*2
upper_bound = mean + std*2

print('Lower Bound :',lower_bound)
print('Upper Bound :',upper_bound)
```
