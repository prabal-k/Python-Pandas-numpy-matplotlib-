### Extracting Sub string from a column using slicing


```python
df6['sort name']=df6['Full Name'].str[0:3]
df6.head()
```

```python
#  in this context of apply(), axis=1 does indeed represent operations along the columns, but it's applied across rows
# This code defines a function age_calc that takes a row (val) as input and returns a string based on whether 
# the value in the 'Age' column of that row is greater than 40 or not. Then, it uses the apply function with
# axis=1 to apply this function to each row of the DataFrame df6, and assigns the result to a new column named 'Value'.
def age_calc1(val):
    if val['Age'] > 40:
        return 'Eligible'
    else:
        return 'Not Eligible'
    
df6['value']=df6.apply(age_calc1,axis=1)
df6.head(10)
```

```python
# This code defines a function age_calc that takes a single value (val) as input and returns a string based on whether
# val is greater than 40 or not. Then, it uses the map function to apply this function to each value in the 
#'Age' column of the DataFrame df6, and assigns the result to a new column named 'Value'.
def age_calc(val):
    if val > 40:
        return 'Eligible'
    else:
        return 'Not Eligible'
    
df6['Value'] = df6['Age'].map(age_calc)
df6.head()
```
