```python
df_c=pd.read_excel(r"C:\Users\Prabal Kuinkel\Desktop\Data Analyst\PYTHON JUPYTERNOOTBOOK\Data sets and notes\Zomato data set\Country-Code.xlsx")
df_c.head(15)
```

```python
# Merging the df (zomato.csv) and df_c (country-code.xlsx) based on common column Country Code
```

```python
final_df=pd.merge(df,df_c,on='Country Code',how='left')
```

```python
final_df.head(10)
```

```python
#All the columns in final_df data set
final_df.columns
```

```python
# Number of unique values is each columns
final_df.nunique()
```

```python
# Count of each countries in Country column
# index          #values
final_df['Country'].value_counts()
```

```python
# Storing the index of Country column in country_name_label
country_name_label=final_df['Country'].value_counts().index
country_name_label
```

```python
# Storing the values of Country column in country_value
country_value=final_df['Country'].value_counts().values
country_value
```

```python
#pie chart to display top 3 countries that uses Zomato
plt.pie(country_value[:3],labels=country_name_label[:3],autopct="%1.1f%%")
```

```python
#Observation from above pie chart
#1. Zomato maximum records or transactions are from India
# followed by United states and United Kingdom
```

```python
final_df.columns
```

```python
final_df.groupby('Country').agg({'Aggregate rating':'mean'}).sort_values(by='Aggregate rating',ascending=False)
```

```python
# The same thing can be done using below method/ beow code ie below 2 cell code
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating counts'})
ratings
```

```python
## conclusion for above:
# When Rating is between 4.5 to 4.9 ------------> indicates excellent
# When Rating is between 4.0 to 4.4 ------------> indicated very Good
# When Rating is between 3.5 to 3.9 ------------> indicated Good
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=ratings,x='Aggregate rating',y='Rating counts')
```

```python
# The same thing can be done using above method/ beow code ie line no 137 and 134
rating=final_df.groupby('Aggregate rating').agg({'Aggregate rating':'count'}).rename(columns={'Aggregate rating':'Rating count'}).reset_index()
rating
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=rating,x='Aggregate rating',y='Rating count')
```
