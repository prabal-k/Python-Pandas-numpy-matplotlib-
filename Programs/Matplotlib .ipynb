```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

# Scatter plot

```python
##scatter plot using matplotlib
plt.scatter(x,y,c='red')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot")
```
